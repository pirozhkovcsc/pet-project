### –ö–æ–ª–æ–Ω–∫–∞ math

–¢–∞–∫ –∫–∞–∫ –∫–æ–ª–æ–Ω–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –º–Ω–æ–≥–æ–∑–Ω–∞—á–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏, —Ç–æ –ª–æ–≥–∏—á–Ω–µ–π –±—ã–ª–æ –±—ã –≤–Ω–µ–¥—Ä–∏—Ç—å –¥–≤–∞ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞. (–Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ –≤ '–î—Ä—É–≥–æ–µ', —É–¥–∞–ª–∏–º, —Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ –≤—ã–±—Ä–æ—Å—ã, —Ç–∞–º –≤—Å–µ–≥–æ –ª–∏—à—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫ –Ω–∞–ø–∏—Å–∞–ª–∏)

```python
math_answers = [
         '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞ - —ç—Ç–æ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–π –∏ –¥—Ä—É–≥–∏—Ö –Ω–∞—É–∫.',
         '–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞ - —ç—Ç–æ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–∞—è –Ω–∞—É–∫–∞, –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –∏–∑—É—á–∞—Ç—å —Ä–∞–¥–∏ —Å–∞–º–æ—Ä–∞–∑–≤–∏—Ç–∏—è.'
               ]

new_features_math = ['apply', 'pure']

for ind, value in zip(new_features_math, math_answers):
    df[ind] = df.math.apply(lambda x: value in x).astype('int')
```

–°—Ç–∞—Ä—É—é —Ñ–∏—á—É –º—ã —É–¥–∞–ª–∏–º, –∞ –¥–≤–∞ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω—É–ª—è –≤ —Å—Ç—Ä–æ–∫–µ –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ —Å –∫–æ–ª–æ–Ω–∫–∞–º–∏ apply –∏ pure –≤—ã–ª–µ–∑–ª–∏ –∏–∑ - –∑–∞ —Ç–æ–≥–æ, —á—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –ø–∏—Å–∞–ª–∏ —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—ã–µ –æ—Ç–≤–µ—Ç—ã. –ò—Ö –±—ã–ª–æ –º–∞–ª–æ, –ø–æ—ç—Ç–æ–º—É –ø—Ä–æ—Å—Ç–æ –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –Ω–µ –±—É–¥–µ–º —Å–ª–µ–¥–∏—Ç—å –∑–∞ —ç—Ç–∏–º–∏ —Å—Ç—Ä–æ—á–∫–∞–º–∏ –≤ —ç—Ç–∏—Ö –¥–≤—É—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö.

```python
df.drop(columns=['math'], inplace=True) # –£–¥–∞–ª–∏–ª–∏ —Å—Ç–∞—Ä—É—é —Ñ–∏—á—É
```

```python
df.subject.unique()
```

```python
df.subject.value_counts() # –¢–µ–æ—Ä–∏—è —Ñ—É–Ω–∫—Ü–∏–π, –≤—ã—Å—à–∞—è –∞–ª–≥–µ–±—Ä–∞ –∏ –º–∞—Ç–ª–æ–≥ - –≤—ã–±—Ä–æ—Å—ã. –û—Ç—Å—é–¥–∞ —ç—Ç–æ –ø—Ä—è–º –æ—á–µ–≤–∏–¥–Ω–æ.
```

```python
subjects = [
        '–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∞—è –≥–µ–æ–º–µ—Ç—Ä–∏—è',
        '–õ–∏–Ω–µ–π–Ω–∞—è –∞–ª–≥–µ–±—Ä–∞',
        '–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑',
        '–î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞'
           ]

new_features_subject = ['sub_angem', 'sub_linal', 'sub_matan', 'sub_discra']

for ind, value in zip(new_features_subject, subjects):
    df[ind] = df.subject.apply(lambda x: value in x).astype('int')
```

### –ö–æ–ª–æ–Ω–∫–∞ useless_subject

(–≤ —ç—Ç–æ–º —Ä–∞–∑–¥–µ–ª–µ –Ω—É–∂–Ω—ã –≤—Å–µ –≥—Ä–∞—Ñ–∏–∫–∏ –∏ –≤–µ–∑–¥–µ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–æ, –ø–æ—á–µ–º—É –∏—Ö –Ω—É–∂–Ω–æ –≤–∑—è—Ç—å)

–ü–æ—Å–º–æ—Ç—Ä–∏–º —Ç–µ–ø–µ—Ä—å –Ω–∞ –∫–æ–ª–æ–Ω–∫—É useless_subject:

–õ–µ–≥–∫–æ –≤–∏–¥–µ—Ç—å, —á—Ç–æ —Ç—É—Ç –Ω—É–∂–Ω–æ –±—ã–ª–æ –¥–æ–±–∞–≤–∏—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã, —Ç–∞–∫ –∫–∞–∫ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞–ª—Å—è –≤–∞—Ä–∏–∞–Ω—Ç **–ù–µ—Ç —Ç–∞–∫–æ–≤—ã—Ö**. –ü—Ä–∏–≤–µ–¥–µ–º –≤ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–∏–¥ —ç—Ç—É –∫–æ–ª–æ–Ω–∫—É, –∞ —Ç–∞–∫–∂–µ –∏–∑–±–∞–≤–∏–º—Å—è –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ (**–∏—Å—Ç–æ—Ä–∏—è**, **–æ–ø–¥**, **–º–∞—Ç–ª–æ–≥** –∏ —Ç.–¥). –í—ã–±—Ä–æ—Å—ã —É–π–¥—É—Ç –≤ –≤–∞—Ä–∏–∞–Ω—Ç **–ù–µ—Ç —Ç–∞–∫–æ–≤—ã—Ö** (—Ç–∞–∫ –∫–∞–∫ –∏—Ö –æ—á–µ–Ω—å –º–∞–ª–æ).

–û—Ç–º–µ—Ç–∏–º —Ç–∞–∫–∂–µ, —á—Ç–æ —ç—Ç–æ—Ç –ø—Ä–∏–∑–Ω–∞–∫ —Ç–∞–∫–∂–µ –∏–º–µ–µ—Ç –º–Ω–æ–≥–æ–∑–Ω–∞—á–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏.

```python
useless_subjects = [
        '–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∞—è –≥–µ–æ–º–µ—Ç—Ä–∏—è',
        '–õ–∏–Ω–µ–π–Ω–∞—è –∞–ª–≥–µ–±—Ä–∞',
        '–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑',
        '–î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞'
                   ]

new_features_useless_subject = ['useless_angem', 'useless_linal', 'useless_matan', 'useless_discra']

for ind, value in zip(new_features_useless_subject, useless_subjects):
    df[ind] = df.useless_subject.apply(lambda x: value in x).astype('int')
```

–£–¥–∞–ª–∏–º —Ñ–∏—á—É `useless_subject`:

```python
df.drop(columns='useless_subject', inplace=True)
```

–î–æ–±–∞–≤–∏–º —Ç–∞–∫–∂–µ —Ñ–∏—á—É useless_none, –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏–µ `True` —Å—Ç–∞–≤–∏—Ç—Å—è, –µ—Å–ª–∏ –≤—Å–µ –ø—Ä–µ–¥–º–µ—Ç—ã –±—ã–ª–∏ –ø–æ–ª–µ–∑–Ω—ã–µ (—Ç.–µ –≤–æ –≤—Å–µ—Ö 4 - –µ—Ö –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö —Å—Ç–æ—è—Ç –Ω—É–ª–∏).

```python
df['useless_none'] = (df.iloc[:, -4:].sum(axis=1) == 0).astype('int')
```

```python
useless_subject_names = df.iloc[:, -5:].columns # –ó–¥–µ—Å—å –±—É–¥—É—Ç –ª–µ–∂–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è –ø—è—Ç–∏ –¥–æ–±–∞–≤–ª–µ–Ω–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
```

–°—Ä–∞–∑—É –≤—ã–≤–µ–¥–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –ø–æ —ç—Ç–æ–π —Ñ–∏—á–µ, –∫–æ—Ç–æ—Ä—É—é —Ç–æ–ª—å–∫–æ —á—Ç–æ —Ä–∞–∑–¥–µ–ª–∏–ª–∏ –Ω–∞ 5. (–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞)

```python
df.groupby(['sex'], as_index=False)[useless_subject_names].sum()
```

–ú–æ–∂–Ω–æ —ç—Ç–æ —Å–¥–µ–ª–∞—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏.

```python
useless_subject_by_sex = df.groupby(['sex'])[useless_subject_names].sum()
useless_subject_by_sex = useless_subject_by_sex.rename(columns={
                  'useless_angem': '–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∞—è –≥–µ–æ–º–µ—Ç—Ä–∏—è',
                  'useless_linal': '–õ–∏–Ω–µ–π–Ω–∞—è –∞–ª–≥–µ–±—Ä–∞',
                  'useless_matan': '–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑',
                  'useless_discra': '–î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞',
                  'useless_none': '–ù–µ—Ç —Ç–∞–∫–æ–≤—ã—Ö'})

res = useless_subject_by_sex.reset_index()
res_wide = res.melt(id_vars='sex')
plt.figure(figsize=(10,8))
ax = sns.barplot(x='sex', y='value',data=res_wide, hue='variable')

ax.set_title('–ù–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–µ –ø—Ä–µ–¥–º–µ—Ç—ã, —Ç–∞–∫ –∫–∞–∫ –Ω–µ—Ç –ø–æ–Ω–∏–º–∞–Ω–∏—è, –∑–∞—á–µ–º –æ–Ω–∏ –Ω—É–∂–Ω—ã (–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞)\n', fontsize=17)
ax.set_xlabel('–ü–æ–ª', fontsize=15)
ax.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ\n', fontsize=15)
ax.legend(title='–í–∞—Ä–∏–∞–Ω—Ç—ã –æ—Ç–≤–µ—Ç–∞',
          ncol=2);
```

–ê —Ç–µ–ø–µ—Ä—å –±–µ–∑ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞.

```python
useless_subject = df[useless_subject_names].sum(axis=0)
fig, ax = plt.subplots(figsize=[12, 8])
labels = useless_subject.keys()
plt.pie(x=useless_subject, autopct="%.1f%%", explode=[0.05]*5, labels=useless_subject_by_sex.columns, pctdistance=0.5)

plt.title("–ù–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–µ –ø—Ä–µ–¥–º–µ—Ç—ã, —Ç–∞–∫ –∫–∞–∫ –Ω–µ—Ç –ø–æ–Ω–∏–º–∞–Ω–∏—è, –∑–∞—á–µ–º –æ–Ω–∏ –Ω—É–∂–Ω—ã", fontsize=17);
```

–û—Ç—Å—é–¥–∞, –∫–∞–∫ –∏ –∏–∑ –ø—Ä–æ—à–ª–æ–π –∫–∞—Ä—Ç–∏–Ω–∫–∏, –≤–∏–¥–Ω–æ, —á—Ç–æ –ª–∏–¥–∏—Ä—É–µ—Ç –∞–Ω–≥–µ–º, –∞ –º–∞—Ç–∞–Ω —É–∂–µ –∞—É—Ç—Å–∞–π–¥–µ—Ä, —Ö–æ—Ç—è –¥–æ —ç—Ç–æ–≥–æ –≤—Å–µ –±—ã–ª–æ –ø–æ - –¥—Ä—É–≥–æ–º—É:)

–û—Ç–º–µ—Ç–∏–º –µ—â–µ –æ–¥–Ω–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–µ:

```python
useless_subject_by_sex = df.groupby(['sex'])[useless_subject_names].mean()
useless_subject_by_sex = useless_subject_by_sex.rename(columns={
                  'useless_angem': '–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∞—è –≥–µ–æ–º–µ—Ç—Ä–∏—è',
                  'useless_linal': '–õ–∏–Ω–µ–π–Ω–∞—è –∞–ª–≥–µ–±—Ä–∞',
                  'useless_matan': '–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑',
                  'useless_discra': '–î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞',
                  'useless_none': '–ù–µ—Ç —Ç–∞–∫–æ–≤—ã—Ö'})

res = useless_subject_by_sex.reset_index()
res_wide = res.melt(id_vars='sex')
plt.figure(figsize=(10,8))
ax = sns.barplot(x='variable', y='value',data=res_wide, hue='sex')

ax.set_title('–ù–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–µ –ø—Ä–µ–¥–º–µ—Ç—ã, —Ç–∞–∫ –∫–∞–∫ –Ω–µ—Ç –ø–æ–Ω–∏–º–∞–Ω–∏—è, –∑–∞—á–µ–º –æ–Ω–∏ –Ω—É–∂–Ω—ã (–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞)\n', fontsize=17)
ax.set_xlabel('–ü–æ–ª', fontsize=15)
ax.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ\n', fontsize=15)
ax.set_xticklabels(labels=useless_subject_by_sex.columns,
                   rotation=15,
                   fontsize=13)
ax.legend(title='–í–∞—Ä–∏–∞–Ω—Ç—ã –æ—Ç–≤–µ—Ç–∞',
          ncol=2);
```
