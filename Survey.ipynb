### Колонка math

Так как колонка содержит многозначные строки, то логичней было бы внедрить два дополнительных признака. (некоторые комментарии, которые были в 'Другое', теперь также будут отражены в отдельном признаке)

Вначале добавим колонки `apply`, `pure` для основных ответов:

```python
math_answers = [
         'Математика - это вспомогательный инструмент для приложений и других наук.',
         'Математика - это самостоятельная наука, которую нужно изучать ради саморазвития.'
               ]

new_features_math = ['apply', 'pure']

for ind, value in zip(new_features_math, math_answers):
    df[ind] = df.math.apply(lambda x: value in x).astype('int')
```

Теперь добавим колонку `other`:

```python
not_other = set(math_answers) 

df['other'] = df.math.apply(lambda x: len(set(x.split(';')).difference(not_other)) > 0).astype('int')
```

Старую фичу мы удалим, а два одновременных нуля в строке на пересечение с колонками apply и pure вылезли из - за того, что некоторые писали развернутые ответы. Их было мало, поэтому просто в дальнейшем не будем следить за этими строчками в этих двух колонках.

```python
df.drop(columns=['math'], inplace=True) # Удалили старую фичу
```

### Колонка useless_subject

(в этом разделе нужны все графики и везде обосновано, почему их нужно взять)

Посмотрим теперь на колонку useless_subject:

```python
df.useless_subject.unique()
```

Легко видеть, что тут нужно было добавить дополнительные варианты, так как часто встречался вариант **Нет таковых**. Приведем в нормальный вид эту колонку, а также избавимся от выбросов (**история**, **опд**, **матлог** и т.д). Выбросы уйдут в вариант **Нет таковых** (так как их очень мало).

Отметим также, что этот признак также имеет многозначные строки.

```python
useless_subjects = [
        'Аналитическая геометрия',
        'Линейная алгебра',
        'Математический анализ',
        'Дискретная математика'
                   ]

new_features_useless_subject = ['useless_angem', 'useless_linal', 'useless_matan', 'useless_discra']

for ind, value in zip(new_features_useless_subject, useless_subjects):
    df[ind] = df.useless_subject.apply(lambda x: value in x).astype('int')
```

Удалим фичу `useless_subject`:

Добавим также фичу useless_none:

```python
not_other = set(useless_subjects) 

df['useless_none'] = df.useless_subject.apply(lambda x: len(set(x.split(';')).difference(not_other)) > 0).astype('int')
```

```python
useless_subject_names = df.iloc[:, -5:].columns # Здесь будут лежать названия пяти добавленных колонок
```

```python
df[useless_subject_names].mean() # Доли
```

```python
df[useless_subject_names].sum() # Количество
```

```python
df.drop(columns='useless_subject', inplace=True) # Удалим колонку
```

Сразу выведем статистику по этой фиче, которую только что разделили на 5. (в зависимости от пола)

```python
df.groupby(['sex'], as_index=False)[useless_subject_names].sum()
```

Можно это сделать графически.

```python
useless_subject_by_sex = df.groupby(['sex'])[useless_subject_names].sum()
useless_subject_by_sex = useless_subject_by_sex.rename(columns={
                  'useless_angem': 'Аналитическая геометрия',
                  'useless_linal': 'Линейная алгебра',
                  'useless_matan': 'Математический анализ',
                  'useless_discra': 'Дискретная математика',
                  'useless_none': 'Нет таковых'})

res = useless_subject_by_sex.reset_index()
res_wide = res.melt(id_vars='sex')
plt.figure(figsize=(10,8))
ax = sns.barplot(x='sex', y='value',data=res_wide, hue='variable')

ax.set_title('Неинтересные предметы, так как нет понимания, зачем они нужны (в зависимости от пола)\n', fontsize=17)
ax.set_xlabel('Пол', fontsize=15)
ax.set_ylabel('Количество\n', fontsize=15)
ax.legend(title='Варианты ответа',
          ncol=2);
```

Заметим, что у нашего лидера "Аналитическая геометрия" бары примерно равны, а значит, зависимости от пола нет (по - хорошему можно использовать критерий, например, хи-квадрат, но здесь это и так очевидно). То есть действительно ангем не понимают вне зависимости от пола, что является уже серьезным звонком.

Заметим также, что в случае с алгеброй процент недовольных девушек в два раза больше (даже больше, чем в два), значит, им желательно больше рассказывать про алгебру. То есть если на спецкурс пришли 5 парней и 2 девушки, то с высокой вероятностью хотя бы одна из них была недовольна алгеброй ($1 - 3/4 \cdot 3/4 = 7/16 \thickapprox 0.5$)

```python
df.groupby(['sex'], as_index=False)[useless_subject_names].mean() # Все величины указаны в процентах!
```

Еще одно замечание: сейчас использовать A/B тесты не имеет смысл. Так как мы еще даже никак не повлияли на студентов. Проведем A/B тестирование уже после спецкурса, это было бы логичным завершением проекта и доказательством того, что мы действительно решили проблему. А ее нужно действительно решить!

Из последних двух вопросов вытекают следующие рассуждения:

Матан люди хотят послушать больше всего и он больше всего понятен. Ангем же хотят послушать чуть меньше, но зато слишком много людей не понимают, зачем он нужен и, как следствие, хотя его из - за этого послушать, как он применяется в жизни. Алгебру же хотят послушать еще меньше, но почему - то у девушек процент непонимания сильно больше, что хотелось бы исправить. Отсюда можно получить 3 основных предмета, которые мы будем обьяснять студентам.
